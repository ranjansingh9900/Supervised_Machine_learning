```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
X,y=make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=20,random_state=13)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression,Ridge
```

```python
lr=LinearRegression()
```

```python
lr.fit(X,y)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
rr=Ridge(alpha=10)
```

```python
rr.fit(X,y)
print(rr.coef_)
print(rr.intercept_)
```

```python
rr1=Ridge(alpha=100)
rr1.fit(X,y)
print(rr1.coef_)
print(rr1.intercept_)
```

```python
plt.plot(X,y,'b.')
plt.plot(X,lr.predict(X),color='red',label='alpha=0')
plt.plot(X,rr.predict(X),color='green',label='alpha=10')
plt.plot(X,rr1.predict(X),color='blue',label='alpha=100')
plt.legend()
```

```python
def linear_regression(X,y,alpha=1):
    x_mean=np.mean(X)
    y_mean=np.mean(y)
    
    nom=0
    dem=0
    
    for i in range(X.shape[0]):
        nom=nom+(y[i]-y_mean)*(X[i]-x_mean)
        dem=dem+(X[i]-x_mean)**2
        
    m=nom/(dem+alpha)
    b=y_mean-m*(x_mean)
    print("m: ",m)
    print("b: ",b)
    
    return m,b
```

```python
linear_regression(X,y)
```
